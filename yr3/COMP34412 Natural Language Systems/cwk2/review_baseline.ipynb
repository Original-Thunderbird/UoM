```python
# note: before running the program, further cleaning on MPQA lexicon should be done:
# line 5549 and 5550: delete 'm' between 'stemmed1=n' and 'priorpolarity=negative';
# line 3749: change 'polarity=negative priorpolarity=weakneg' to 'priorpolarity=negative'
# Then format of info of each word is normalized and reading can be conveniently done.
from nltk import word_tokenize
```

```python
# do classification task
def baseline(source, corpus):
    f = open(source, "r", encoding='latin-1')
    raw = f.read().split('\n')[:-1]
    corpus_pol = source[-3:]
    count = 0 # for current corpus, how many examples are correctly classified
    for rev in raw:
        pol = 0
        words = word_tokenize(rev)
        # sum up polarities of all words from MPQR to give final polarity of the example
        for word in words:
            if word in dictionary:
                pol += dictionary[word]
        if pol > 0:
            corpus.append((words, 'positive'))
            if corpus_pol == 'pos':
                count += 1
        elif pol == 0:
            corpus.append((words, 'neutral'))
        else:
            corpus.append((words, 'negative'))
            if corpus_pol == 'neg':
                count += 1
    return count
```

```python
# baseline
pol_to_val_map = {'positive':1, 'negative':-1, 'both':0, 'neutral':0}

# load dictionary
f = open('subjectivity_clues_hltemnlp05/subjclueslen1-HLTEMNLP05.tff', "r", encoding='utf-8')
dictionary = {}
raw = f.read().split('\n')[:-1]
for line in raw:
    bar = line.split()
    word = bar[2][6:]
    # key:word -> value:strength
    if word not in dictionary:
        dictionary[word] = pol_to_val_map[bar[5][14:]]

for word, info in dictionary.items():
    print(f'{word} {info}')
print(len(dictionary))
```

```python
neg_corpus = []
neg_count = baseline('corpus2/rt-polarity.neg', neg_corpus)
neg_acc = neg_count/len(neg_corpus)
pos_corpus = []
pos_count = baseline('corpus2/rt-polarity.pos', pos_corpus)
pos_acc = pos_count/len(pos_corpus)
tot_acc = (neg_count+pos_count)/(len(neg_corpus)+len(pos_corpus))

print('pos_corpus:')
for corpora in pos_corpus:
    print(corpora)
print('neg_corpus:')
for corpora in neg_corpus:
    print(corpora)
print('pos_corpus accuracy: '+str(pos_acc))
print('neg_corpus accuracy: '+str(neg_acc))
print('overall accuracy: '+str(tot_acc))
```
